# Scans used for reference, taken at B=0.5T
Very similar to the previous plots but with slightly higher resolution. To plot the boxes, you need to convert the boxes given in voltages to pixel values, use the helper function for this

```python
import conversions

data_dir = "experiment_2/data/label_and_ref_scans"
file_name = "reference_scan_positive bias_magnet_0.5.pkl"
scan_path = os.path.join(data_dir, file_name)
with open(scan_path, "rb") as f:
    scan_data = pickle.load(f)
voltage_boxes = np.load("experiment_2/data/boxes/boxes_positive_bias_voltage_space.npy")
px_boxes = conversions.convert_vboxes_to_pxboxes(voltage_boxes, scan_data)

print("Scan:")
plotting_utils.draw_in_voltagespace(scan_data)

print("With all boxes:")
img = scan_data["chan0"]["data"]
condition = "positive bias"
plotting_utils.draw_boxes_in_img_pxspace(img, px_boxes, condition)


file_name = "reference_scan_negative bias_magnet_0.5.pkl"
scan_path = os.path.join(data_dir, file_name)
with open(scan_path, "rb") as f:
    scan_data = pickle.load(f)
voltage_boxes = np.load("experiment_2/data/boxes/boxes_negative_bias_voltage_space.npy")
px_boxes = conversions.convert_vboxes_to_pxboxes(voltage_boxes, scan_data)

print("Scan:")
plotting_utils.draw_in_voltagespace(scan_data)

print("With all boxes:")
img = scan_data["chan0"]["data"]
condition = "negative bias"
plotting_utils.draw_boxes_in_img_pxspace(img, px_boxes, condition)
```

# Measurements at different magnetic fields
## Positive bias

```python
data_dir = "experiment_2/data/positive bias/"
for file_name in os.listdir(data_dir):
    if file_name[-4:] == ".pkl":
        scan_path = os.path.join(data_dir, file_name)
        with open(scan_path, "rb") as f:
            scan_data = pickle.load(f)
        magnetic_field = float(file_name.split("_")[-1][:-4])
        print(f"magnetic_field={np.round(magnetic_field,3)}T")
        plotting_utils.draw_in_voltagespace(scan_data)
```

## Confirmation scans

```python
data_dir = "experiment_2/data/confirmation_scans/"
for conf_scan_path in [
    "confirmation_scan_second_run_box_0.pkl",
    "confirmation_scan_second_run_box_1.pkl",
    "confirmation_scan_second_run_box_2.pkl",
]:
    with open(os.path.join(data_dir,conf_scan_path), "rb") as f:
        confirmation_scan = pickle.load(f)

    magnet_vals = np.array(confirmation_scan["magnet_range"])
    start_detuning_point = confirmation_scan["start_detuning_point"]
    end_detuning_point = confirmation_scan["end_detuning_point"]
    # resolution = confirmation_scan['resolution']
    distance = np.sqrt(np.sum((end_detuning_point - start_detuning_point) ** 2))
    # detuning_vals= np.array([t*(end_detuning_point-start_detuning_point)+start_detuning_point
    #             for t in np.linspace(0,1, int(distance/resolution))])
    scan = np.array(confirmation_scan["scan"])

    # The zero point of the detuning axis needs to be manually shifted to recover the plots from the publication
    # Additionally, the plots in the publication are cropped to the relevant region
    plt.imshow(
        scan.T[::-1],
        cmap="icefire",
        extent=[magnet_vals.min(), magnet_vals.max(), 0, distance],
        aspect=0.01,
    )
    plt.xlabel("Magnetic field in T")
    plt.ylabel("Detuning in mV")
    plt.show()
```
