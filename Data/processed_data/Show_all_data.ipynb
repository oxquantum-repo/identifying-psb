````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns # needed for 'icefire' colormap
````

````python
imgs = np.load('finfets_imgs.npy', allow_pickle=True)
labels = np.load('finfets_labels.npy', allow_pickle=True)
data_set_names = np.load('finfets_names.npy', allow_pickle=True)
device_names = np.load('finfets_device_names.npy', allow_pickle=True)
````

```img``` contains a pair of stability diagrams, taken with a potentially blocked (index 0) and unblocked condition (index 1)

```labels``` contains the label of whether there is PSB present or not

```data_set_names``` gives information on where the raw data is stored

```device_names``` gives the name of the device used

````python
len(imgs)
````

````python
for img, label, name, device in zip(imgs,labels,data_set_names,device_names):
    
    fig, axs = plt.subplots(2,1)
    axs[0].imshow(img[0], origin='lower', cmap='icefire')
    axs[0].set_title(f'PSB: {bool(label)}\nname: {name}\ndevice name: {device}\n\npotentially blocked')
    axs[1].imshow(img[1], origin='lower', cmap='icefire')
    axs[1].set_title('unblocked')
    plt.tight_layout()
    plt.show()
    print('*'*50)
````
